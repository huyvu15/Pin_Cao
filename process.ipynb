```python
import pandas as pd

# Đọc dữ liệu từ file CSV
file_path = 'anime1.csv'  # Thay thế bằng đường dẫn tệp của bạn
df = pd.read_csv(file_path)

# Xử lý cột 'follower'
def convert_follower(follower):
    follower = follower.replace('followers', '').replace(' ', '').replace(',', '')
    try:
        if 'k' in follower:
            return int(float(follower.replace('k', '')) * 1000)
        elif 'M' in follower:
            return int(float(follower.replace('M', '')) * 1000000)
        else:
            return int(follower)
    except ValueError:
        return 0

df['follower'] = df['follower'].apply(convert_follower)

# Chuyển đổi tất cả các giá trị trong cột 'comment_count' thành chuỗi để xử lý
df['comment_count'] = df['comment_count'].astype(str)

# Xóa hết "comments" trong cột comment_count
df['comment_count'] = df['comment_count'].str.replace('Comments', '').str.replace('Comment', '').str.strip()

# Thay thế giá trị trống và NaN trong cột 'comment_count' bằng 0
df['comment_count'] = df['comment_count'].replace(['', 'nan'], '0').apply(lambda x: int(float(x)))

# Chuyển hết "Author not found" trong cột author thành None
df['author'] = df['author'].replace('Author not found', None)

# Lưu lại file CSV mới sau khi xử lý
cleaned_file_path = 'anime_data.csv'  # Thay thế bằng đường dẫn tệp lưu trữ
df.to_csv(cleaned_file_path, index=False)

print("Dữ liệu đã được xử lý và lưu tại:", cleaned_file_path)
```
